```python
data1=df.drop(columns=['gender','race/ethnicity','parental level of education','test preparation course','lunch'] , axis=1)
plt.figure(figsize=(20,5))
for i,col in enumerate(data1.columns):
    plt.subplot(1,3,i+1)
    sns.boxplot( x=col  , data=data1)
    plt.xlim(-10,110)
```

```python
total=((df['math score'] + df['reading score'] + df['writing score'])/300)*100
df['total score']=total 
df
```
